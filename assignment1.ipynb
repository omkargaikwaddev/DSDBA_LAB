```python
data = df['Addicted_Score']
```

```python
print("Mean:", np.mean(data))
print("Std:", np.std(data))
print("Min:", np.min(data))
print("Max:", np.max(data))
```

```python
mean = data.mean()
std = data.std()
```

```python
threshold = 3
outliers = []

for i in data:
    z_score = (i - mean) / std
    if np.abs(z_score) > threshold:
        outliers.append(i)

outliers
```
